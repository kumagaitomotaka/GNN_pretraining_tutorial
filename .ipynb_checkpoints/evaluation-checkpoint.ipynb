```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,roc_auc_score, mean_squared_error, mean_absolute_error,roc_curve, auc
from sklearn.metrics import r2_score
import os
import pickle
from sklearn.preprocessing import label_binarize

def learning_curve(file_dir):
    loss_file = os.path.join(file_dir, 'logs.csv')
    fig,ax = plt.subplots(figsize=(6,4))
    loss_df = pd.read_csv(loss_file)
    ax.plot(loss_df['epoch'], loss_df['train_loss'], label = 'Train')
    ax.plot(loss_df['epoch'], loss_df['val_loss'], label = 'Validation')
    #タイトル
    ax.set_title('Loss')
    # 凡例を表示
    ax.legend(loc=1)
    # 軸ラベル
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Loss')
    
def QM_rgr_eval(file_dir, n='all'):
    PandL_file = os.path.join(file_dir, 'PandL.npz')
    npz = np.load(PandL_file)
    if n == 'all':
        predictions = []
        labels = [] 
        preds = []
        y = []
        for i in range(2):
            predictions.extend(npz['pred'][i])
            preds.append(npz['pred'][i])
            labels.extend(npz['labels'][i])
            y.append(npz['labels'][i])
    else:
        predictions = npz['pred'][n]
        labels = npz['labels'][n]
    R2 = r2_score(labels, predictions) 
    #Prediction Error Plot
    fig = plt.figure(figsize=(6,6))
    if n == 'all':
        plt.scatter(y[0], preds[0], s=10, alpha=0.6, label = 'HOMO')
        plt.scatter(y[1], preds[1], s=10, alpha=0.6, label = 'LUMO')
    else:
        plt.scatter(labels, predictions, s=10, alpha=0.6)
    plt.plot([-1, 1], [-1, 1], color = "red")
    if n == 0:
        plt.title('{} Prediction'.format('HOMO'))    # 図のタイトル
    elif n == 1:
        plt.title('{} Prediction'.format('LUMO'))    # 図のタイトル
    else:
        plt.title('Prediction')    # 図のタイトル
    plt.ylabel('Predict') # y軸のラベル
    plt.xlabel('True Value')    # x軸のラベル
    if n == 0:
        plt.text(min(labels)-min(labels)*0.1,max(labels)+max(labels)*0.1,'R2: {:.2f}'.format(R2))
    elif n == 1:
        plt.text(min(labels)-min(labels)*0.1,max(labels)-max(labels)*0.1,'R2: {:.2f}'.format(R2))
    elif n == 'all':
        h_r2 = r2_score(y[0], preds[0]) 
        l_r2 = r2_score(y[1], preds[1])
        plt.text(min(labels)-min(labels)*0.1,max(labels)-max(labels)*0.5,' R2(all): {:.2f} \n R2(HOMO): {:.2f} \n R2(LUMO): {:.2f}'.format(R2, h_r2, l_r2))
    plt.grid()  
    plt.xlim([min(labels), max(labels)])
    plt.ylim([min(labels), max(labels)])
    if n == 'all': plt.legend(loc='lower right')
    plt.show() 
    
def evaluation(file_dir):
    PandL_file = os.path.join(file_dir, 'PandL.npz')
    npz = np.load(PandL_file)
    pred = npz['pred']
    labels = npz['labels']
    pred = np.argmax(pred, axis=1)
    matrix_fp = confusion_matrix(labels, pred)
    plt.figure()
    sns.set(rc = {'figure.figsize':(8,8)})
    splot = sns.heatmap(matrix_fp, square=True, annot=True, cbar=False, fmt='d', cmap='GnBu')
    sfig = splot.get_figure()
    accuracy = accuracy_score(labels, pred)
    precision_list = precision_score(labels, pred, average=None)
    recall_list = recall_score(labels, pred, average=None)
    print('Accuracy(正解率): {:.3f}'.format(accuracy))
    print('macro-Precision(適合率): {:.3f}'.format(np.average(precision_list)))
    print('macro-Recall(再現率): {:.3f}'.format(np.average(recall_list)))
    return accuracy, precision_list, recall_list

def roc_auc(file_dir):
    PandL_file = os.path.join(file_dir, 'PandL.npz')
    npz = np.load(PandL_file)
    pred = npz['pred']
    labels = npz['labels']
    fpr, tpr, _ = roc_curve(labels, pred[:, 1])
    roc_auc = auc(fpr, tpr)
    print('AUC: {:.3f}'.format(auc(fpr, tpr)))
    plt.plot(fpr, tpr)
    plt.xlabel('FPR: False positive rate')
    plt.ylabel('TPR: True positive rate')
    
def roc_auc_mult(file_dir):
    PandL_file = os.path.join(file_dir, 'PandL.npz')
    npz = np.load(PandL_file)
    pred = npz['pred']
    labels = npz['labels']
    n_classes = len(pred[0])
    label_class = np.array(range(n_classes))
    one_hot_labels = label_binarize(labels, classes=label_class)
    fpr = {}
    tpr = {}
    roc_auc = {}
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(one_hot_labels[:, i], pred[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
        print('AUC_{}: {:.3f}'.format(i, auc(fpr[i], tpr[i])))
    for i in range(n_classes):
        plt.plot(fpr[i], tpr[i], label=f'class: {i}')
        plt.xlabel('FPR: False positive rate')
        plt.ylabel('TPR: True positive rate')
    plt.legend()
    print('AUC_All: {:.3f}'.format(sum(roc_auc.values())/len(roc_auc)))
    
def rgr_eval(file_dir):
    PandL_file = os.path.join(file_dir, 'PandL.npz')
    npz = np.load(PandL_file)
    predictions = npz['pred']
    labels = npz['labels']
    R2 = r2_score(labels, predictions) 
    #Prediction Error Plot
    fig = plt.figure(figsize=(6,6))
    plt.scatter(labels, predictions, s=10, alpha=0.6)
    plt.plot([-10, 2], [-10, 2], color = "red")
    plt.title('Yields Prediction(%)')    # 図のタイトル
    plt.ylabel('Predict') # y軸のラベル
    plt.xlabel('True Value')    # x軸のラベル
    plt.text(-9,1.5,'R2: {:.2f}'.format(R2))
    plt.grid()  
    #plt.xlim([-10, 2])
    #plt.ylim([-10, 2])
    plt.show() 
```
